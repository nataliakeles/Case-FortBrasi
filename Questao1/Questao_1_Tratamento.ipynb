```python
df_nulos = df[df['estimated_population_2019'].isna() == True]
df_nulos
```

Como estão distribuidos os dados faltantes? Será que é de forma aleatória entre os estados, ou todos os dados faltantes se concentram no mesmo estado e podem influenciar na análise?

```python
df_agg = df.groupby('state').count()
df_nulo_agg = df_nulos.groupby('state').count()
state = df_nulo_agg.index
```

```python
Abaixo verificamos o número de mortes e casos confirmados que serão desconsiderados casos sejam excluídas as linhas com dados fal
```

```python
df_nulo_sum = df_nulos.groupby('state').count()
df_nulo_sum
```

```python
df_faltantes_UF = pd.DataFrame(columns = ['UF','Total de registros', 'Total de registros faltantes', 'Resgistros faltantes %'])
```

```python
for uf in state:
    tr = df_agg.loc[uf, 'date']
    tr_faltantes = df_nulo_agg.loc[uf, 'date']
    tr_faltantes_percent = (df_nulo_agg.loc[uf, 'date']/df_agg.loc[uf, 'date'])*100
    df_faltantes_UF.loc[len(df_faltantes_UF)] = [uf, tr, tr_faltantes, tr_faltantes_percent]
```
